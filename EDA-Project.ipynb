```python
def waterfront(waterfront):
    if waterfront==0 :
        return 'no'
    else:
        return 'yes'


df['waterfront'] = df['waterfront'].apply(waterfront)
```

```python
def renovate(year_renovated):
    if year_renovated ==0 :
        return 'no'
    else:
        return 'yes'


df['renovate'] = df['year_renovated'].apply(renovate)
```
